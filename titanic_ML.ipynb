```python
import pandas as pd

titanic_train_data=pd.read_csv("train.csv")
titanic_train_data.head()
```

```python
titanic_test_data=pd.read_csv("test.csv")
titanic_test_data.head()
```

```python
titanic_train_data.isnull().sum()
```

```python
import numpy as np

mean=np.mean(titanic_train_data['Age'])
titanic_train_data['Age']=titanic_train_data['Age'].fillna(mean)
```

```python
titanic_train_data['Cabin']=titanic_train_data['Cabin'].fillna(method='bfill')
```

```python
titanic_train_data['Embarked']=titanic_train_data['Embarked'].fillna(method='bfill')
```

```python
titanic_train_data.isnull().sum()
```

```python
titanic_test_data.isnull().sum()
```

```python
mean=np.mean(titanic_test_data['Age'])
titanic_test_data['Age']=titanic_test_data['Age'].fillna(mean)
```

```python
titanic_test_data['Cabin']=titanic_test_data['Cabin'].fillna(method='bfill')
titanic_test_data['Fare']=titanic_test_data['Fare'].fillna(method='ffill')
```

```python
titanic_test_data.isnull().sum()
```

# Model Initiation

```python
titanic_test_data.head()
```

```python
titanic_train_data.head()
```

```python
y_target=titanic_train_data.Survived

X_train=pd.get_dummies(titanic_train_data[['Pclass','Sex','SibSp','Parch']])
X_test=pd.get_dummies(titanic_test_data[['Pclass','Sex','SibSp','Parch']])
```

```python
from sklearn.ensemble import RandomForestClassifier as RFC

model_titanic=RFC(n_estimators=100, max_depth=5, random_state=1)
model_titanic.fit(X_train, y_target)

#Make prediction
predict_titanic=model_titanic.predict(X_test)
```

```python
predicted_passenger_survived=pd.DataFrame({"PassengerID":titanic_test_data.PassengerId,
                                           "Survived":predict_titanic
                                          }
                                         )
predicted_passenger_survived
```
